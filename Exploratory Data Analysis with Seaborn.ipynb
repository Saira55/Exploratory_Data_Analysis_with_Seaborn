```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
data=pd.read_csv('C:\\Users\\Uzma\\j\\Projects\\1. Exploratory Data Analysis with Seaborn\\data.csv')
```

```python
data.head()
```

```python
col = data.columns
print(col)
```

```python
y=data.diagnosis

drop_col=['Unnamed: 32','id','diagnosis']
x=data.drop(drop_col, axis=1)
x.head()
```

```python
ax=sns.countplot(y=y, label="count")
B,M=y.value_counts()
print("Number of Benign Tumors", B)
print("Number of Malignant Tumors", M)
```

```python
x.describe()
```

```python
data = x
data_std=(data-data.mean())/data.std()
data= pd.concat([y, data_std.iloc[:, 0:10]], axis=1)
data=pd.melt(data, id_vars="diagnosis",
            var_name="features",
            value_name="value")

plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=data, split=True, inner="quart")
plt.xticks(rotation=45);
```
